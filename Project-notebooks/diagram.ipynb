```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data for the charts
shipping_modes = ['Standard Class', 'Second Class', 'First Class', 'Same Day']
shipping_counts = [60, 20, 15, 5]

customer_segments = ['Consumer', 'Corporate', 'Home Office']
customer_counts = [5000, 3000, 2000]

product_categories = ['Office Supplies', 'Furniture', 'Technology']
product_sales = [6000, 2000, 2000]

# Creating a figure to hold all the subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Shipping Modes Pie Chart
axs[0].pie(shipping_counts, labels=shipping_modes, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
axs[0].set_title('Shipping Modes')

# Customer Segments Bar Chart
sns.barplot(x=customer_segments, y=customer_counts, palette='pastel', ax=axs[1])
axs[1].set_title('Customer Segments')
axs[1].set_ylabel('Count')

# Product Categories Bar Chart
sns.barplot(x=product_categories, y=product_sales, palette='pastel', ax=axs[2])
axs[2].set_title('Product Categories')
axs[2].set_ylabel('Sales')

# Adjust layout
plt.tight_layout()

# Save the figure
diagram_path = "./E-Commerce_Sales_Analysis_Simple.png"
plt.savefig(diagram_path)

plt.show()

diagram_path
```

```python


# Data for the diagram
ship_modes = ['Standard Class', 'Second Class', 'First Class', 'Same Day']
ship_values = [5968, 1945, 1538, 543]

segments = ['Consumer', 'Corporate', 'Home Office']
segment_values = [5191, 3020, 1783]

categories = ['Office Supplies', 'Furniture', 'Technology']
category_sales = [6026, 2121, 1847]
category_profit = [40.0, 15.0, 45.0]  # Example profit percentages

states = ['California', 'New York', 'Texas']
state_sales = [2001, 1128, 985]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Shipping Modes Pie Chart
axs[0, 0].pie(ship_values, labels=ship_modes, autopct='%1.1f%%', startangle=140)
axs[0, 0].set_title('Shipping Modes Sales Distribution')

# Customer Segments Bar Plot
sns.barplot(x=segments, y=segment_values, ax=axs[0, 1])
axs[0, 1].set_title('Customer Segments Distribution')
axs[0, 1].set_ylabel('Number of Customers')

# Product Categories Sales and Profit
sns.barplot(x=categories, y=category_sales, ax=axs[1, 0], color='blue', alpha=0.6, label='Sales')
sns.barplot(x=categories, y=category_profit, ax=axs[1, 0], color='red', alpha=0.6, label='Profit %')
axs[1, 0].set_title('Product Categories: Sales & Profit')
axs[1, 0].legend()

# Top Performing States Bar Plot
sns.barplot(x=states, y=state_sales, ax=axs[1, 1])
axs[1, 1].set_title('Top Performing States')
axs[1, 1].set_ylabel('Number of Sales')

# Adjust layout
plt.tight_layout()

# Save the diagram
plt.savefig('/mnt/data/ECommerce_Sales_Analysis_Diagram.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data (to be replaced with actual data from the project)
shipping_modes = pd.DataFrame({
    'Mode': ['Standard Class', 'Second Class', 'First Class', 'Same Day'],
    'Sales': [5968, 1945, 1538, 543]
})

customer_segments = pd.DataFrame({
    'Segment': ['Consumer', 'Corporate', 'Home Office'],
    'Sales': [5191, 3020, 1783]
})

product_categories = pd.DataFrame({
    'Category': ['Office Supplies', 'Furniture', 'Technology'],
    'Sales': [6026, 2121, 1847]
})

yearly_trends = pd.DataFrame({
    'Year': [2019, 2020, 2021, 2022],
    'Sales': [1993, 2102, 2587, 3312]
})

profit_margins = pd.DataFrame({
    'Category': ['Office Supplies', 'Furniture', 'Technology'],
    'Profit': [30000, 15000, 50000]
})

geo_distribution = pd.DataFrame({
    'State': ['California', 'New York', 'Texas', 'Pennsylvania', 'Washington'],
    'Sales': [2001, 1128, 985, 587, 506]
})

fig, axs = plt.subplots(3, 2, figsize=(14, 18))
fig.suptitle('E-Commerce Sales Analysis', fontsize=20, weight='bold')

# Shipping Modes Pie Chart
axs[0, 0].pie(shipping_modes['Sales'], labels=shipping_modes['Mode'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
axs[0, 0].set_title('Sales by Shipping Mode')

# Customer Segments Bar Plot
sns.barplot(x='Segment', y='Sales', data=customer_segments, palette='viridis', ax=axs[0, 1])
axs[0, 1].set_title('Sales by Customer Segment')

# Product Categories Bar Plot
sns.barplot(x='Category', y='Sales', data=product_categories, palette='muted', ax=axs[1, 0])
axs[1, 0].set_title('Sales by Product Category')

# Yearly Trends Line Plot
sns.lineplot(x='Year', y='Sales', data=yearly_trends, marker='o', ax=axs[1, 1])
axs[1, 1].set_title('Yearly Sales Trends')

# Profit Margins Bar Plot
sns.barplot(x='Category', y='Profit', data=profit_margins, palette='coolwarm', ax=axs[2, 0])
axs[2, 0].set_title('Profit Margins by Category')

# Geographic Distribution Bar Plot
sns.barplot(x='Sales', y='State', data=geo_distribution, palette='Blues_d', ax=axs[2, 1])
axs[2, 1].set_title('Sales by State')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.savefig('./E_Commerce_Sales_Analysis_Infographic.png')
plt.show()
```
